Wheather data was selected as a correlated variable and thus fitted with symbolic regression.

This symbolic regression uses an iterative algorithm to find linear and non-linear coefficients. 

Instead of the relation $Ax = b$ a more complex combination of unknown ($x$) is obtained.

The symbolic regression algorithm was obtained from the `SybolicRegressor` object in the `gplearn` library.

```python
import gplearn as gpl
from gplearn.genetic import SymbolicRegressor,SymbolicTransformer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score,precision_score,recall_score)
import datetime
```

```python
data = pd.read_excel("Consumo.xlsx")
```

```python
sns.set()
```

- A set of minimum, maximum, average, cloudiness, rain, and wind speed variables was created from weighted sums (Wheater conditions from multiple municipalities).

- For this set, no correlation proof was computed.

```python
data["mes"]=data.fecha.apply(lambda x :x.month)
data.head()
```

## Monthly data

- Categorical variables of each month and maximum temperature were used as independent variables. 

- Sum, subtraction, multiplication, and sinusoidal operations/functions were added to the algorithm as possible coefficients for the fitted model.

```python
data1 = data[["mes","max"]]
data1.head()
```

```python
# get dummies
data_dummies = pd.get_dummies(data1,columns=["mes"])
data_dummies.head()
```

```python
X = data_dummies.drop("max",axis=1).values
y = data_dummies["max"].values
```

```python
y
```

```python
# add+, sub-, mul*, sin, cos
est = SymbolicRegressor(function_set=('add', 'sub', 'mul','sin','cos'))
est.fit(X,y)
```

```python
# combination of unknowns
print(est._program)
```

```python
y_pred = est.predict(X)
```

```python
# Plot time (x), vs temperature (y)
plt.figure(figsize=(10,8))
plt.plot(y)
plt.plot(y_pred)
```

## With train/test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
est = SymbolicRegressor(function_set=('add', 'sub', 'mul','sin','cos'))
est.fit(X_train,y_train)
```

```python
print(est._program)
```

```python
# Prediction for all the dataset
y_pred = est.predict(X)
```

```python
plt.figure(figsize=(15,10))
plt.plot(y, label="Real")
plt.plot(y_pred, label="Predicción")
plt.legend()
```

```python
# Prediction for the test
y_pred = est.predict(X_test)
```

```python
plt.figure(figsize=(15,10))
plt.plot(y_test, label="Real")
plt.plot(y_pred, label="Predicción")
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```

```python
np.mean(np.abs((y_test-y_pred)/y_test))*100
```

# monthly and weekly categorical variables

- For this case, the week of the year was added as categorical variables in the independent variable set.

```python
data["week"]=data.fecha.apply(lambda x: datetime.date(x.year, x.month, x.day).isocalendar()[1])
data3=data[["max","mes","week"]]
data3.head()
```

```python
data3_dummies=pd.get_dummies(data3,columns=["mes","week"])
data3_dummies.head()
```

```python
X = data3_dummies.drop("max",axis=1).values
y = data3_dummies["max"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
est = SymbolicRegressor(function_set=( 'mul','sin','cos',"add","sub"))
est.fit(X_train,y_train)
```

```python
print(est._program)
```

```python
# Prediction with monthly and weekly variables (all set)
y_pred = est.predict(X)
plt.figure(figsize=(15,10))
plt.plot(y, label="Real")
plt.plot(y_pred, label="Predicción")
plt.legend()
```

```python
y_pred = est.predict(X_test)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```

```python
np.mean(np.abs((y_test-y_pred)/y_test))*100
```
